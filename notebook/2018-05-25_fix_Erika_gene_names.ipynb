Erika provided a list of genes for annotation in a spreadsheet. Unfortunately, the gene names are not exact flybase gene symbols. I needed to go through and normalize these names for woking in the workflows.

```python
def check_genes(genes):
    for gene in genes:
        try:
            nbconfig.symbol2fbgn[gene]
        except KeyError:
            print(gene)
```

 # Y Expression Stats

 Quick stats for Y expression by cluster.

```python
import os
import pandas as pd

from larval_gonad.stats import run_chisq

try:
    os.chdir(os.path.join(os.getcwd(), "notebook"))
    print(os.getcwd())
except:
    pass
```

```python
fbgn2chrom = (
    pd.read_feather(
        "../references/gene_annotation_dmel_r6-26.feather", columns=["FBgn", "FB_chrom"]
    )
    .set_index("FBgn")
    .squeeze()
    .rename("chrom")
)
```

```python
df = pd.read_feather("../output/seurat3-cluster-wf/tpm_by_cluster.feather").set_index("FBgn").join(
    fbgn2chrom
).set_index(["chrom", "cluster"], append=True).unstack().query("chrom == 'Y'") # type: pd.DataFrame
df.columns = df.columns.droplevel()
```

```python
ct = (df > 0).apply(lambda x: x.value_counts(), axis=0)
ct.columns = ct.columns.astype(str)
```

```python
res = run_chisq(ct)
```

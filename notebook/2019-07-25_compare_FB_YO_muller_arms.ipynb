 # Muller Element Comparison FB vs YO

 I have ortholog lists from FlyBase and Haiwang's data. I want to go through
 and comapre the Muller element calls to see how much disagreement there is.

```python
import os
import pandas as pd
from sklearn.metrics import adjusted_rand_score

try:
    os.chdir(os.path.join(os.getcwd(), 'notebook'))
    print(os.getcwd())
except:
    pass
```

```python
fb = pd.read_feather('../output/neox-wf/FB_muller_arm_assignment.feather').set_index('FBgn').fillna("NAN")
yo = pd.read_feather('../output/neox-wf/YO_muller_arm_assignment.feather').set_index('FBgn').fillna("NAN")
```

```python
common_fbgns = fb.index.intersection(yo.index)
print(f"Haiwang: {yo.shape[0]:,}")
print(f"FlyBase: {fb.shape[0]:,}")
print(f'There are {common_fbgns.shape[0]:,} genes in common between Haiwang and FlyBase')
```

```python
fb = fb.reindex(common_fbgns)
yo = yo.reindex(common_fbgns)
```

```python
def compare(species):
    score = adjusted_rand_score(fb[species], yo[species])
    return species, score
```

```python
pd.DataFrame((compare(species) for species in fb.columns), columns=['species', 'adj_rand']).set_index("species")
```
